### Fazendo a conversão

```python
import gtts
from playsound import playsound

tts = gtts.gTTS("Oi Mundo", lang="pt")
tts.save("./arquivos/oi_mundo.mp3")
playsound("./arquivos/oi_mundo.mp3")
```

### Listando as linguagens disponíveis

### Fazendo a conversão

```python
import pyttsx3

robo = pyttsx3.init()
robo.setProperty('rate', 90)
robo.setProperty('voice', 'brazil')
message = "Essa é uma mensagem de teste"#input("Escreva o que quer que eu diga: ")

robo.say(message)
robo.runAndWait()
```

### Listando as linguagens disponíveis

```python
voices = robo.getProperty('voices')
for voice in voices:
    print("Voice:", voice.name)
    print(" - ID: " , voice.id)
    print(" - Languages: " , voice.languages)
    print(" - Gender: " , voice.gender)
    print(" - Age: " , voice.age)
    print("\n")
```
