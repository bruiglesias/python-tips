### Pegando o indice bovespa

```python
from pandas_datareader import data as web
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# formato da data MM-DD-YYYY
# exemplos de ticker:
# - ^BVSP  (bovespa)
# - MGLU3.SA (magazine luiza)
cotacao_bovespa = web.DataReader('^BVSP', data_source='yahoo', 
                                 start="01/01/2020", end="01/01/2021")
display(cotacao_bovespa)
```

```python
cotacao_bovespa['Adj Close'].plot(figsize=(15, 10))
plt.show()
```

### Pegando a cotação de uma lista de empresas

```python
tabela_empresas = pd.read_excel("arquivos/Empresas.xlsx")
display(tabela_empresas)
```

```python
for empresa in tabela_empresas['Empresas']:
    print(empresa)
    cotacao = web.DataReader(f'{empresa}.SA', data_source='yahoo', 
                                 start="01/01/2020", end="01/01/2021")
    display(cotacao)
    cotacao['Adj Close'].plot(figsize=(15, 10))
    plt.show()
```
